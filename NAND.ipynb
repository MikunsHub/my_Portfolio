```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
```

```python
# The input and output, i.e. truth table, of a NAND gate
x_train = np.array([[0,0],[0,1],[1,0],[1,1]], "uint8")
y_train = np.array([[1],[1],[1],[0]], "uint8")
```

```python
# Create neural networks model
model = Sequential()
# Add layers to the model
model.add(Dense(output_dim=4, activation='relu', input_dim=2))      # first input layer
model.add(Dense(output_dim=4, activation='relu'))                   # second hidden layer
model.add(Dense(output_dim=1, activation='sigmoid'))                # output layer
```

```python
# Compile the neural networks model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# Train the neural networks model
model.fit(x_train, y_train, nb_epoch=7000)
```

```python
# Test the output of the trained neural networks based NAND gate
y_predict = model.predict(x_train)
print(np.round(y_predict))
```

```python
y_pred = model.predict(x_train)
#Converting predictions to label
pred = list()
for i in range(len(y_pred)):
    pred.append(np.argmax(y_pred[i]))
#Converting one hot encoded test label to label
test = list()
for i in range(len(y_train)):
    test.append(np.argmax(y_train[i]))
```

```python
from sklearn.metrics import accuracy_score
a = accuracy_score(pred,test)
print('Accuracy is:', a*100)
```

```python
# save model as h5 file
model.save("NAND_model.h5")
```
